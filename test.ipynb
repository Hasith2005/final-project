```python
from pysr import PySRRegressor
import numpy as np

X = 2 * np.random.randn(100, 5)
y = 2 * np.cos(X[:, 4]) + X[:, 0] ** 2 - 2
model = PySRRegressor(binary_operators=["+", "-", "*", "/"],unary_operators=["sin","cos","tan"])
model.fit(X, y)
print(model)
```

```python
model.sympy() # wow, it got it righttt!!!
```
