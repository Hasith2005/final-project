```python
import pysr
import pandas as pd
```

```python
ds_eq1 = pd.read_csv('feynman-bonus.1.txt',delimiter=' ')
ds_eq1.columns = ['Z1','Z2','KE','t','A']
ds_eq1.head()
```

```python
y=ds_eq1.iloc[:,-1]
X=ds_eq1.iloc[:,:4]
```

```python
from pysr import PySRRegressor

model=PySRRegressor(binary_operators=["+","-","/","*"],unary_operators=['square','cube','inv'])
X_units = ["kgm^2s^-1","ms^-1","kgm^2s^-2"]
y_units=["m^2"]
model.fit(X,y)
```

```python
best_score=model.equations_.sort_values("score",axis=0,ascending=False).iloc[[0]]
best_score
```
